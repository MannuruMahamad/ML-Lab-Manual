```python
#information gain 
def find_entropy(df):

    target = df.columns[-1]

    values = df[target].unique()

    entropy = 0

    for value in values:

        probability = len(df[df[target] == value]) / len(df)

        entropy += -probability * np.log2(probability)

    return entropy
```

```python
def find_entropy_attribute(df, attribute):

    target = df.columns[-1]

    values = df[attribute].unique()

    total_entropy = 0

    for value in values:

        subset = df[df[attribute] == value]

        subset_entropy = 0

        target_values = subset[target].unique()

        for target_value in target_values:

            probability = len(subset[subset[target] == target_value]) / len(subset)

            subset_entropy += -probability * np.log2(probability)

        weight = len(subset) / len(df)

        total_entropy += weight * subset_entropy

    return total_entropy
```

```python
# split information for Gain Ratio 
import numpy as np

def split_information(df, attribute):
    values = df[attribute].unique()
    split_info = 0

    for value in values:
        probability = len(df[df[attribute] == value]) / len(df)

        if probability > 0:
            split_info += -probability * np.log2(probability)

    return split_info
```

```python
#Gain ratio for selecting best attribute which fits as a splitting criteria
def gain_ratio(df, attribute):
    info_gain = find_entropy(df) - find_entropy_attribute(df, attribute)

    split_info = split_information(df, attribute)

    if split_info == 0:
        return 0

    return info_gain / split_info
```

```python
# finding the gini index for CART 
def find_gini(df):
    target = df.columns[-1]

    values = df[target].unique()
    gini=1.0
    for value in values:
        probability=len(df[df[target]==value])/len(df)
        gini -= probability**2
    return gini
    
    
```

```python
# wweighted Gini impurity for a specific tuple or attribute
def find_gini_attribute(df,attribute):
    values=df[attribute].unique()
    total_gini = 0
    for value in values:
        subset=df[df[attribute] == value]
        subset_gini = find_gini(subset)
        weight = len(subset)/len(df)
        total_gini += weight * subset_gini
    return total_gini
```

```python
#find winner for c4.5
def find_winner_c45(df):

    GR = []

    attributes = []

    for key in df.columns[:-1]:

        # Ignore identifier columns like RollNo
        if df[key].nunique() == len(df):
            continue

        GR.append(gain_ratio(df, key))
        attributes.append(key)

    return attributes[np.argmax(GR)]
```

```python
#find the best split vakue attribute for IG3
def find_winner(df):
    IG = []
    attributes = []

    # Ignore the target column (last column)
    for key in df.columns[:-1]:

        # Skip identifier columns like RollNo
        if df[key].nunique() == len(df):
            continue

        gain = find_entropy(df) - find_entropy_attribute(df, key)
        IG.append(gain)
        attributes.append(key)

    return attributes[np.argmax(IG)]
```

```python
def find_winner_cart(df):
    Gini_scores=[]
    attributes=[]
    for key in df.columns[:-1]:
        #skip identifier columns like rollno
        if df[key].nunique() == len(df):
            continue
        Gini_scores.append(find_gini_attribute(df,key))
        attributes.append(key)
    return attributes[np.argmin(Gini_scores)]
```

```python
# trying to fit suitable attribute
def get_subtable(df,attribute,value):
    return df[df[attribute] == value].reset_index(drop = True)
    
```

```python
def buildtree(df,tree=None):
    node = find_winner(df)
    attvalue=np.unique(df[node])
    Class=df.keys()[-1]
    if tree is None:
        tree={}
        tree[node]={}
    for value in attvalue:
        subtable = get_subtable(df,node,value)
        C1value,counts=np.unique(subtable[Class],return_counts=True)
        if len(counts)==1:
            tree[node][value]=C1value[0]
        else:
            tree[node][value]=buildtree(subtable)
    return tree
```

```python
def buildtreec_45(df,tree=None):
    node = find_winner_c45(df)
    attvalue=np.unique(df[node])
    Class=df.keys()[-1]
    if tree is None:
        tree={}
        tree[node]={}
    for value in attvalue:
        subtable = get_subtable(df,node,value)
        C1value,counts=np.unique(subtable[Class],return_counts=True)
        if len(counts)==1:
            tree[node][value]=C1value[0]
        else:
            tree[node][value]=buildtreec_45(subtable)
    return tree
```

```python
# Tree builder for CART matching your current function style
def buildtree_cart(df, tree=None):
    node = find_winner_cart(df)
    attvalue = np.unique(df[node])
    Class = df.columns[-1]
    
    if tree is None:
        tree = {}
        tree[node] = {}
        
    for value in attvalue:
        subtable = get_subtable(df, node, value)
        C1value, counts = np.unique(subtable[Class], return_counts=True)
        
        if len(counts) == 1:
            tree[node][value] = C1value[0]
        else:
            tree[node][value] = buildtree_cart(subtable)
            
    return tree
```
