```python
import pandas as pd

# Load Any Dataset

file_name = input("Enter CSV file name: ")
df = pd.read_csv(file_name)

print("\nDataset:")
print(df)

# Target Column (Last Column)

target = df.columns[-1]

# Enter Test Sample

print("\nEnter values for the following features:")

X = {}
for feature in df.columns[:-1]:
    value = input(f"{feature}: ")
    X[feature] = value

# Naive Bayes Classification
classes = df[target].unique()

posterior_probs = {}


print("\nNaive Bayes Calculation")


for cls in classes:

    class_df = df[df[target] == cls]

    # Prior Probability
    prior = len(class_df) / len(df)

    print(f"\nClass = {cls}")
    print(f"P({cls}) = {len(class_df)}/{len(df)} = {prior:.4f}")

    likelihood = 1

    # Conditional Probabilities
    for feature, value in X.items():

        count = len(class_df[class_df[feature] == value])

        # Laplace Smoothing
        unique_values = df[feature].nunique()
        prob = (count + 1) / (len(class_df) + unique_values)

        likelihood *= prob

        print(f"P({feature}={value}|{cls}) = ({count}+1)/({len(class_df)}+{unique_values}) = {prob:.4f}")

    posterior = prior * likelihood
    posterior_probs[cls] = posterior

    print(f"P(X|{cls}) = {likelihood:.8f}")
    print(f"Posterior = {posterior:.8f}")


# Normalize Probabilities

total = sum(posterior_probs.values())

print("\nPosterior Probabilities")

for cls in posterior_probs:
    posterior_probs[cls] /= total
    print(f"P({cls}|X) = {posterior_probs[cls]:.6f}")

# -----------------------------
# Prediction
# -----------------------------
prediction = max(posterior_probs, key=posterior_probs.get)


print("Prediction")


print("\nTest Sample:")
for feature, value in X.items():
    print(f"{feature} : {value}")

print(f"\nPredicted Class = {prediction}")
```
